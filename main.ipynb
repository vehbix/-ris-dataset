```python
from keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.impute import SimpleImputer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

SVC ile iris

```python
# Load the iris dataset
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# Veri setini eğitim ve test setlerine böl
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Replace missing values with the mean
imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)
```

```python
# Özellikleri ölçeklendirin
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_imputed)
X_test_scaled = scaler.transform(X_test_imputed)
```

```python
# Yapay zeka modelini oluşturun ve eğitin
model = SVC()
model.fit(X_train_scaled, y_train)
```

```python
# Modelin performansını değerlendirin
accuracy = model.score(X_test_scaled, y_test)
print("Model accuracy:", accuracy)
```

Decision Tree Clasifier ile iris

```python
# Create a decision tree classifier
clf = DecisionTreeClassifier()
```

```python
# Train the classifier
clf.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = clf.predict(X_test)
```

```python
# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# Calculate the confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)
```

knn ile iris

```python
# Create a KNN classifier
knn = KNeighborsClassifier()
```

```python
# Train the classifier
knn.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred_knn = knn.predict(X_test)
```

```python
# Calculate the accuracy of the KNN model
accuracy_knn = accuracy_score(y_test, y_pred_knn)
print("KNN Accuracy:", accuracy_knn)
```

```python
# Calculate the confusion matrix for KNN
cm_knn = confusion_matrix(y_test, y_pred_knn)
print("KNN Confusion Matrix:")
print(cm_knn)
```

Sepuential model ile iris

```python
# Etiketleri bir-hot kodlamasına çevir
y_s = to_categorical(y)
y_s.shape
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train_s, y_test_s = train_test_split(X, y_s, test_size=0.3, random_state=42)
```

```python
X_train.shape
```

```python
# Modeli tanımla
model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(y_s.shape[1], activation='softmax'))  
```

```python
# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
model.fit(X_train, y_train_s, epochs=100, batch_size=512, validation_data=(X_test, y_test_s))
```

```python
# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Test loss: {loss:.4f}')
print(f'Test accuracy: {accuracy:.4f}')
```
